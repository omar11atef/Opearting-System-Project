# FCFS

```python
class FCFSAlgorithm:
    def run(self):
        try:
            process = int(input("Enter Number of Processes: "))

            top = 0
            at = []
            bt = []
            ct = []
            tat = []
            wt = []
            for i in range(process):
                arrival_time = float(input(f"Enter Arrival Time for P{i}: "))
                burst_time = float(input(f"Enter Burst Time for P{i}: "))
                at.append(arrival_time)
                bt.append(burst_time)

                # calc Completion Time
                if top < at[i]:
                    top = at[i] + bt[i]
                else:
                    top += bt[i]
                ct.append(top)

                # calc TAT and WT
                tat.append(ct[i] - at[i])
                wt.append(tat[i] - bt[i])

            print("\nCompletion Time calculated...")
            
            #Print Table Elements :
            print("\nPID\tAT\tBT\tCT\tTAT\tWT")
            for i in range(process):
                print(f"{i+1}\t{at[i]}\t{bt[i]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")

            avgTAT = sum(tat) / process
            avgWT = sum(wt) / process

            print(f"\nAverage Turn around Time = {avgTAT}")
            print(f"Average Waiting Time = {avgWT}")
        except ValueError:
            print("Error: Please enter a valid number")
fcfs = FCFSAlgorithm()
fcfs.run()
```

# SJFPreemptive

```python
class SJFPreemptive:
    def run(self):
        try:
            process = int(input("Enter Number of Processes: "))

            at = []  # Arrival Time
            bt = []  # Burst Time
            rt = []  # Remaining Time
            for i in range(process):
                arrival_time = float(input(f"Enter Arrival Time for P{i}: "))
                burst_time = float(input(f"Enter Burst Time for P{i}: "))
                at.append(arrival_time)
                bt.append(burst_time)
            rt = bt.copy()  # Remaining time initially = Burst Time

            ct = [0] * process  # Completion Time
            tat = [0] * process  # Turnaround Time
            wt = [0] * process  # Waiting Time

            completed = 0
            current_time = 0
            min_rt = float('inf')
            shortest = -1
            check = False

            while completed != process:

                min_rt = float('inf')
                shortest = -1
                check = False
                for i in range(process):
                    if at[i] <= current_time and rt[i] > 0:
                        if rt[i] < min_rt:
                            min_rt = rt[i]
                            shortest = i
                            check = True
                        elif rt[i] == min_rt:
                            if at[i] < at[shortest]:
                                shortest = i
                if not check:
                    current_time += 1
                    continue

            
                rt[shortest] -= 1
                current_time += 1

             
                if rt[shortest] == 0:
                    completed += 1
                    ct[shortest] = current_time
                    tat[shortest] = ct[shortest] - at[shortest]
                    wt[shortest] = tat[shortest] - bt[shortest]

            print("\nCompletion Time calculated...")
            print("\nPID\tAT\tBT\tCT\tTAT\tWT")
            for i in range(process):
                print(f"{i+1}\t{at[i]}\t{bt[i]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")

            avgTAT = sum(tat) / process
            avgWT = sum(wt) / process

            print(f"\nAverage Turn around Time = {avgTAT}")
            print(f"Average Waiting Time = {avgWT}")

        except ValueError:
            print("Error: Please enter a valid number")
sjf = SJFPreemptive()
sjf.run()
```

# SJF Non-Preemptive

```python
class SJFNonPreemptive:
    def run(self):
        try:
            process = int(input("Enter Number of Processes: "))

            at = []  # Arrival Time
            bt = []  # Burst Time
            for i in range(process):
                arrival_time = float(input(f"Enter Arrival Time for P{i}: "))
                burst_time = float(input(f"Enter Burst Time for P{i}: "))
                at.append(arrival_time)
                bt.append(burst_time)

            ct = [0] * process  # Completion Time
            tat = [0] * process  # Turnaround Time
            wt = [0] * process  # Waiting Time
            completed = [False] * process  # to check process
            current_time = 0
            completed_count = 0

            while completed_count < process:
                # Find Short Way :
                idx = -1
                min_bt = float('inf')
                for i in range(process):
                    if at[i] <= current_time and not completed[i]:
                        if bt[i] < min_bt:
                            min_bt = bt[i]
                            idx = i
                        elif bt[i] == min_bt:  # Brust are equals , select min Arrival
                            if at[i] < at[idx]:
                                idx = i
                if idx == -1:
                    current_time += 1  # Not ready
                    continue

                # Exteced all 
                current_time += bt[idx]
                ct[idx] = current_time
                tat[idx] = ct[idx] - at[idx]
                wt[idx] = tat[idx] - bt[idx]
                completed[idx] = True
                completed_count += 1

            # Print Elements
            print("\nCompletion Time calculated...")
            print("\nPID\tAT\tBT\tCT\tTAT\tWT")
            for i in range(process):
                print(f"{i+1}\t{at[i]}\t{bt[i]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")

            avgTAT = sum(tat) / process
            avgWT = sum(wt) / process

            print(f"\nAverage Turn around Time = {avgTAT}")
            print(f"Average Waiting Time = {avgWT}")

        except ValueError:
            print("Error: Please enter a valid number")
sjf_non_preemptive = SJFNonPreemptive()
sjf_non_preemptive.run()
```

#  Priority Non-Preemptive

```python
def priority_non_preemptive():
    try:
        process = int(input("Enter Number of Processes: "))

        at = []         
        bt = []         
        priority = []  
        ct = [0] * process
        tat = [0] * process
        wt = [0] * process
        finished = [False] * process

        # Read Process
        for i in range(process):
            at.append(float(input(f"Enter Arrival Time for P{i}: ")))
            bt.append(float(input(f"Enter Burst Time for P{i}: ")))
            priority.append(float(input(f"Enter Priority for P{i}: ")))

        current_time = 0
        completed = 0
        # Select the highest-priority process that has already arrived
        while completed < process:
            idx = -1
            best_priority = float("inf")

            # Find best priority
            for i in range(process):
                if not finished[i] and at[i] <= current_time:
                    if priority[i] < best_priority:
                        best_priority = priority[i]
                        idx = i

            # nothing arrived yet → go to the next arrival time
            if idx == -1:
                current_time += 1
                continue

            # Run the selected process
            current_time += bt[idx]
            ct[idx] = current_time
            tat[idx] = ct[idx] - at[idx]
            wt[idx] = tat[idx] - bt[idx]

            finished[idx] = True
            completed += 1

        # Print Results
        print("\nPID\tAT\tBT\tPr\tCT\tTAT\tWT")
        for i in range(process):
            print(f"{i+1}\t{at[i]}\t{bt[i]}\t{priority[i]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")

        avgTAT = sum(tat) / process
        avgWT = sum(wt) / process

        print(f"\nAverage Turnaround Time = {avgTAT}")
        print(f"Average Waiting Time = {avgWT}")

    except ValueError:
        print("Error: Please enter a valid number")
priority_non_preemptive()
```

#  Priority Preemptive: 

```python
def priority_preemptive():
    try:
        process = int(input("Enter Number of Processes: "))
        at = []         # Arrival Time
        bt = []         # Burst Time
        rt = []         # Remaining Time
        priority = []   # Priority (lower value = higher priority)
        ct = [0] * process
        tat = [0] * process
        wt = [0] * process
        for i in range(process):
            at.append(float(input(f"Enter Arrival Time for P{i}: ")))
            bt.append(float(input(f"Enter Burst Time for P{i}: ")))
            priority.append(float(input(f"Enter Priority for P{i}: ")))
            rt.append(bt[i])  # remaining time initially = burst time

        completed = 0
        current_time = 0

        while completed < process:
            idx = -1
            best_priority = float("inf")

            # Choose process with smallest priority among arrived ones
            for i in range(process):
                if at[i] <= current_time and rt[i] > 0:
                    if priority[i] < best_priority:
                        best_priority = priority[i]
                        idx = i

            # If no process has arrived yet → CPU idle
            if idx == -1:
                current_time += 1
                continue

            # Run process for 1 unit (Preemptive)
            rt[idx] -= 1
            current_time += 1

            # If process finishes → calculate CT, TAT, WT
            if rt[idx] == 0:
                ct[idx] = current_time
                tat[idx] = ct[idx] - at[idx]
                wt[idx] = tat[idx] - bt[idx]
                completed += 1
                
        # Print Results
        print("\nPID\tAT\tBT\tPr\tCT\tTAT\tWT")
        for i in range(process):
            print(f"{i+1}\t{at[i]}\t{bt[i]}\t{priority[i]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")

        avgTAT = sum(tat) / process
        avgWT = sum(wt) / process

        print(f"\nAverage Turnaround Time = {avgTAT}")
        print(f"Average Waiting Time = {avgWT}")

    except ValueError:
        print("Error: Please enter a valid number")
priority_preemptive()
    
```

# Round Robin

```python
class RoundRobinAlgorithm:
    def run(self):
        try:
            n = int(input("Enter Number of Processes: "))
            at = []
            bt = []
            pid = []
            for i in range(n):
                arrival_time = float(input(f"Enter Arrival Time for P{i+1}: "))
                burst_time = float(input(f"Enter Burst Time for P{i+1}: "))
                at.append(arrival_time)
                bt.append(burst_time)
                pid.append(f"P{i+1}")

            quantum = float(input("Enter Quantum Time: "))

            rem_bt = bt.copy()
            ct = [0] * n
            tat = [0] * n
            wt = [0] * n
            time = 0
            queue = []
            visited = [False] * n
            completed = 0

            # Add first arrived processes
            for i in range(n):
                if at[i] == 0:
                    queue.append(i)
                    visited[i] = True
            if not queue:
                first = at.index(min(at))
                time = at[first]
                queue.append(first)
                visited[first] = True

            # Round Robin Loop
            while completed < n:
                if not queue:
                    time += 1
                    for i in range(n):
                        if at[i] <= time and not visited[i]:
                            queue.append(i)
                            visited[i] = True
                    continue

                idx = queue.pop(0)

                if rem_bt[idx] > quantum:
                    time += quantum
                    rem_bt[idx] -= quantum
                else:
                    time += rem_bt[idx]
                    rem_bt[idx] = 0
                    ct[idx] = time
                    completed += 1

                for i in range(n):
                    if at[i] <= time and not visited[i]:
                        queue.append(i)
                        visited[i] = True

                if rem_bt[idx] > 0:
                    queue.append(idx)

            # Calculate TAT and WT
            for i in range(n):
                tat[i] = ct[i] - at[i]
                wt[i] = tat[i] - bt[i]

            # Print Table
            print("\nPID\tAT\tBT\tCT\tTAT\tWT")
            for i in range(n):
                print(f"{pid[i]}\t{at[i]}\t{bt[i]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")

            avgTAT = sum(tat) / n
            avgWT = sum(wt) / n
            print(f"\nAverage Turnaround Time = {avgTAT}")
            print(f"Average Waiting Time = {avgWT}")

        except ValueError:
            print("Error: Please enter a valid number")

rr = RoundRobinAlgorithm()
rr.run()
```

# Multilevel Queue :

```python
class MultilevelQueueAlgorithm:
    def run(self):
        try:
            n = int(input("Enter Number of Processes: "))
            pid = []
            at = []
            bt = []
            qn = []   # Queue Number
            ct = []
            tat = []
            wt = []

            for i in range(n):
                print(f"\nProcess P{i+1}")
                pid.append(f"P{i+1}")
                at.append(float(input("Enter Arrival Time: ")))
                bt.append(float(input("Enter Burst Time: ")))
                qn.append(int(input("Enter Queue Number (1 or 2): ")))

            time = 0

            # Execute Queue 1 (High Priority)
            print("\nExecuting Queue 1 (FCFS)...")
            for i in range(n):
                if qn[i] == 1:
                    if time < at[i]:
                        time = at[i]
                    ct.append(time + bt[i])
                    time = ct[-1]
                    tat.append(ct[-1] - at[i])
                    wt.append(tat[-1] - bt[i])

            # Execute Queue 2 (Lower Priority)
            print("\nExecuting Queue 2 (FCFS)...")
            for i in range(n):
                if qn[i] == 2:
                    if time < at[i]:
                        time = at[i]
                    ct.append(time + bt[i])
                    time = ct[-1]
                    tat.append(ct[-1] - at[i])
                    wt.append(tat[-1] - bt[i])

            # Print Table
            print("\nPID\tQ\tAT\tBT\tCT\tTAT\tWT")
            index = 0
            total_tat = 0
            total_wt = 0

            for i in range(n):
                print(f"{pid[i]}\t{qn[i]}\t{at[i]}\t{bt[i]}\t{ct[index]}\t{tat[index]}\t{wt[index]}")
                total_tat += tat[index]
                total_wt += wt[index]
                index += 1

            avgTAT = total_tat / n
            avgWT = total_wt / n

            print(f"\nAverage Turnaround Time = {avgTAT}")
            print(f"Average Waiting Time = {avgWT}")

        except ValueError:
            print("Error: Please enter valid numbers")


mlq = MultilevelQueueAlgorithm()
mlq.run()
```
